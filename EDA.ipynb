```python
#Check that fix worked
count = 0
rows = 0

for i, text in enumerate(df['Text']):
    if text == "Error fetching article" or (not pd.notna(text)):
        count += 1
    if pd.notna(text):
        flag = True
        words = re.findall(r'\S+', str(text))
        for word in words:
            if any(ord(char) > 127 for char in word):
                if(flag):
                    flag = False
                    rows += 1
                print(f"Row {i+1}: {word}")

print(f"Total empty text: {count}")
print(f"Total invalid text: {rows}")
```

```python
def checkASCII(text):
    if pd.isna(text):
        return False
    
    text = str(text)
    return any(ord(char) > 127 for char in text) or text == "Error fetching article" or not pd.notna(text)

filtered_df = df[
    (~df['Text'].apply(checkASCII))
].copy().dropna()

count = 0
rows = 0

for i, text in enumerate(filtered_df['Text']):
    if text == "Error fetching article" or (not pd.notna(text)):
        count += 1
    if pd.notna(text):
        flag = True
        words = re.findall(r'\S+', str(text))
        for word in words:
            if any(ord(char) > 127 for char in word):
                if(flag):
                    flag = False
                    rows += 1
                print(f"Row {i+1}: {word}")

print(f"Total empty text: {count}")
print(f"Total invalid text: {rows}")
print(len(filtered_df))
filtered_df.to_csv("filtered.csv", index=False)
```
