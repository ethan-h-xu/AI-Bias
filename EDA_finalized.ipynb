```python
#Import Environment
import pandas as pd
from ftfy import fix_text
import re
df = pd.read_csv('Political_Bias.csv')
print(df.head)
print(len(df))
```

```python
#Drop NA
print(df['Text'].isna().sum())
df = df.dropna(subset=['Text'])
print(df['Text'].isna().sum())
```

```python
#Check Bias
bias_counts = df['Bias'].value_counts()
print(bias_counts)

print(f"\nTotal records: {len(df)}")
print(f"Unique categories: {df['Bias'].nunique()}")
print(f"Missing values: {df['Bias'].isnull().sum()}")
```

```python
#Fix Encoding & Emojis
rows = 0
for i, text in enumerate(df['Text']):
    if pd.notna(text):
        flag = True
        words = re.findall(r'\S+', str(text))
        for word in words:
            if any(ord(char) > 127 for char in word):
                if(flag):
                    flag = False
                    rows += 1
                
print(f"Total invalid text before: {rows}")

def removeEmojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "\U0001F900-\U0001F9FF"  # supplemental symbols and pictographs
        "\U00002600-\U000026FF"  # miscellaneous symbols
        "\U00002700-\U000027BF"  # dingbats
        "\U0000FE00-\U0000FE0F"  # variation selectors
        "\U0001F018-\U0001F270"  # various symbols
        "\U00010000-\U0010FFFF"  # other high plane emojis
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

df['Title'] = df['Title'].astype(str).apply(fix_text)
df['Text'] = df['Text'].astype(str).apply(fix_text)
df['Text'] = df['Text'].apply(removeEmojis)

rows = 0
for i, text in enumerate(df['Text']):
    if pd.notna(text):
        flag = True
        words = re.findall(r'\S+', str(text))
        for word in words:
            if any(ord(char) > 127 for char in word):
                if(flag):
                    flag = False
                    rows += 1
                print(f"Row {i+1}: {word}")
                
print(f"Total invalid text after: {rows}")
```

```python
df.to_csv("new_filtered.csv")
```
